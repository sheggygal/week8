```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10, 10, 400)
```

```python
y = x**2
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='blue', linewidth=2, label='y = x^2')

plt.title('Plot of y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.legend()

plt.tight_layout()
plt.show()
```

```python
products = ['A', 'B', 'C', 'D']
sales = [15, 30, 45, 20]
```

```python
plt.figure(figsize=(8, 5))
plt.bar(products, sales, color='skyblue')

plt.title('Weekly Sales Data of Products')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.ylim(0, 50)

for i, v in enumerate(sales):
    plt.text(i, v + 1, str(v), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [40, 30, 20, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
```

```python
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

plt.legend(labels, loc='best', bbox_to_anchor=(1, 0.5))

plt.title('Favorite Fruits')

plt.axis('equal')

plt.tight_layout()
plt.show()
```
